```python
import numpy as np
import matplotlib.pyplot as plt
E = np.load("E.npy")
x = np.arange(len(E))

plt.plot(x, E[:, 0], label="Kinetic Energy")
plt.plot(x, E[:, 1], label="Elastic Energy")
plt.plot(x, E[:, 2], label="Gravitational Energy")
plt.plot(x, E[:, 0] + E[:, 1] + E[:, 2], label="Total Energy")
plt.xlabel("Frame")
plt.ylabel("Energy")
plt.title("Energy over Time")
plt.legend()
plt.show()
```
