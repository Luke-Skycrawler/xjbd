```python
import numpy as np
from g2m.medial import SlabMesh

def scale_slabmesh(nv, scale):
    sm = SlabMesh(f"assets/bunny/ma/bunny_v{nv}.ma")
    sm.V *= scale
    sm.R *= scale
    f = f"assets/bunny/ma/out/bunny_v{nv}.ma"

    sm.export_ma(f)
for nv in [40, 60, 70, 80, 90, 100, 150, 200]:
    scale_slabmesh(nv, 1e-3)
```
